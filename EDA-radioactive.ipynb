```python
reduced_dd.columns=new
#cambiar el nombre de las columnas
reduced_dd.head()
reduced_dd.drop("measuring_target",axis=1,inplace=True)
#1 es columna 0 es fila,inplace modifica el original y no realiza una copia del archivo
```

```python
reduced_dd.loc[reduced_dd["134cs_radioactivity_concentration_"]=="ND","134cs_radioactivity_concentration_"]=0

reduced_dd["134cs_radioactivity_concentration_"]=reduced_dd["134cs_radioactivity_concentration_"].astype(float)
#cambiar tipo de dato astype
reduced_dd.describe().T
def around_changer(x):
    try:
        x=float(x)
        return(x)
    except:
        if isinstance(x,float) or isinstance(x,int) or x==None:
            return x
        elif "around" in x:
                try:
                    x=float(x.split(" ")[1])
                    return x
                except:
                    x=x.split(" ")[1]
        if "〜"in x:
                vals= x.split("〜")
                x=(float(vals[1])+float(vals[0]))/2
        elif '-' in x:
                vals = x.split("-")
                x= (float(vals[1])+float(vals[0]))/2
        elif '^' in x or '(' in x: 
                return None
                return x
        
            

for col in reduced_dd.columns[10:]:
    reduced_dd.loc[reduced_dd[col]=='ND',col]=0
    reduced_dd.loc[reduced_dd[col]=='-',col]=None
    reduced_dd[col]=reduced_dd[col].apply(around_changer)




stat = []
for columna in reduced_dd:
    if "statistical_error" in columna:
        stat.append(columna)
#axis= 1 quiere decir columnas
reduced_dd = reduced_dd.drop(stat,axis=1)

reduced_dd.describe()

#cambiar a formato de fecha

reduced_dd
 
```

```python
reduced_dd.sampling_date = pd.to_datetime(reduced_dd.sampling_date)
reduced_dd.correction_base_date = pd.to_datetime(reduced_dd.correction_base_date)



def time (x):
    if isinstance (x,float):
        return x
    vals = x.split(":")
    return float(vals[0]+vals[1])/60

reduced_dd.sampling_time=reduced_dd.sampling_time.apply(time)

reduced_dd.describe().T

reduced_dd.to_csv("Radioactive data")

```
